```python
import math

epsilon = 0.1647
m1, m3 = 0.130, 4.010
alpha_M = -0.14829
M_Pl_eV = 1.22e28

# Track 1: Ghost condition check
# For scalar-tensor theory L = (M_Pl^2/2 + xi*phi^2)*R
# Ghost-free condition: xi < 1/6
# Our coupling: xi = sqrt(m3/m1) * epsilon^2 / M_Pl

coupling_raw = math.sqrt(m3/m1) * epsilon**2
xi = coupling_raw / M_Pl_eV

print("TRACK 1: GHOST CONDITION CHECK")
print(f"sqrt(m3/m1):       {math.sqrt(m3/m1):.4f}")
print(f"epsilon^2:         {epsilon**2:.6f}")
print(f"coupling (raw):    {coupling_raw:.6f}")
print(f"xi = coupling/MPl: {xi:.4e}")
print(f"Ghost limit:       1/6 = {1/6:.6f}")
print(f"Ghost-free:        {xi < 1/6}")
print()

# Track 2: 3 generations x 10 dof check
N_gen = 3
dof_per_gen = 10  # quark doublet x3 colors + lepton doublet + singlets
N_total = N_gen * dof_per_gen
coeff_track2 = math.sqrt(N_total)

print("TRACK 2: SYMMETRY CHECK")
print(f"3 generations x 10 dof = {N_total}")
print(f"sqrt(30) = {coeff_track2:.5f}")
print(f"Target   = 5.4700")
print(f"Off by   = {abs(coeff_track2-5.47)/5.47*100:.3f}%")
print()

# Seesaw mass check
m_dirac = math.sqrt(m1 * m3)
print("SEESAW STRUCTURE")
print(f"m1 (light):  {m1:.3f} eV")
print(f"m_Dirac:     {m_dirac:.4f} eV")
print(f"m3 (heavy):  {m3:.3f} eV")
print(f"Ratio m3/m_Dirac: {m3/m_dirac:.4f}")
print(f"Ratio m_Dirac/m1: {m_dirac/m1:.4f}")
print(f"Seesaw: m_light = m_Dirac^2/m_heavy = {m_dirac**2/m3:.4f} eV")
print(f"Check: should equal m1 = {m1}")
print(f"Exact seesaw: {abs(m_dirac**2/m3 - m1) < 0.001}")
print()
print("VERDICT:")
if xi < 1/6:
    print("Track 1: Ghost-free. Lagrangian is physically stable.")
else:
    print("Track 1: Ghost instability. Lagrangian fails.")
print(f"Track 2: sqrt(30) is {abs(coeff_track2-5.47)/5.47*100:.3f}% from target.")
print(f"Seesaw: m1 and m3 satisfy exact seesaw relation: {abs(m_dirac**2/m3 - m1) < 0.001}")
```
