```python
import math

# Does the factor 5.47 come from the decay physics?
# Gamma = 150 km/s/Mpc, f = 0.023, g = 1.04e-5
g = 1.04e-5
f = 0.023
factor_decay = 1 / (g * f)

# Or from the number of e-folds?
N_efolds = 60   # standard inflation
factor_efold = N_efolds / (4 * math.pi)

# Or from the daughter mass ratio?
m1, m2, m3 = 0.130, 0.320, 4.010
factor_mass = m3 / (m1 + m2)

print(f"Target factor:        5.4685")
print(f"1/(g*f):              {factor_decay:.2f}")
print(f"N_efolds/(4*pi):      {factor_efold:.2f}")
print(f"m3/(m1+m2):           {factor_mass:.4f}")
print(f"m3/m1:                {m3/m1:.4f}")
print()
print("Closest match:", min([
    (abs(factor_decay - 5.4685), '1/(g*f)'),
    (abs(factor_efold - 5.4685), 'N/(4pi)'),
    (abs(factor_mass - 5.4685), 'm3/(m1+m2)'),
    (abs(m3/m1 - 5.4685), 'm3/m1')
], key=lambda x: x[0]))
```
